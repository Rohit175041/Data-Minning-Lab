```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("collage100.csv")
df.head()
```

```python
unwanted_char=['%','$','*','/',',']
def rem_unwanted_admin_rate(item):
  data=item.loc['Admission Rate']
  for character in unwanted_char:
    if(character=='%'):
      data=data.replace('%','')

  if(data=='NA'):
    return np.NaN
  return data

df['Admission Rate'] = df.apply(rem_unwanted_admin_rate, axis = 1)
df.head()
```

```python
def rem_unwanted_sta_act(item):
  data=str(item.loc['*SAT or ACT'])
  for character in unwanted_char:
    if(character=='*'):
      data=data.replace('*','')
    if(character=='%'):
      data=data.replace('%','')
    
  if(data=='NA'):
    return np.NaN
  S = int(data[:data.find('/')])
  A = int(data[data.find('/')+1:])
  marks = max(S,A)
  return str(marks)

df['*SAT or ACT'] = df.apply(rem_unwanted_sta_act, axis = 1)
df.head()
```
