```python
import numpy as np
import pandas as pd
```

```python
# read csv data
db=pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\sem 6th\DataMiLab\impfile\OnlineRetail.csv")
print(db.head())
# for row in db:
#     print(row)
```

```python
# 1. Find mean, std dev, min , max of unitprice field.
print("min:",db["UnitPrice"].min()," Std dev:",db["UnitPrice"].std()," max:",db["UnitPrice"].max()," mean:",db["UnitPrice"].mean())
```

```python
# 2. Draw histogram of unitprice field using matplotlib
import numpy as np 
import matplotlib.pyplot as plt
plt.hist(db["UnitPrice"])
plt.title('Histogram') 
# plt.ylabel('number') 
plt.xlabel('UnitPrice')
plt.axis([-12062.06, 40000.0,0,8])
plt.show()
```

```python
# 3. Find mean of unitprice group by country and plot them
unitprice=['UnitPrice']
dataset=db.groupby('Country')[unitprice].mean()
# print(dataset)
plt.title('mean of unitprice group by country')
plt.plot(dataset)
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Find min and max of unit price group by customerID
min = db.groupby('CustomerID')["UnitPrice"].min()
max = db.groupby('CustomerID')["UnitPrice"].max()
plt.figure(figsize=(20, 5))
plt.subplot(121)
plt.title('min of unitprice group by customerID')
plt.plot(min)
plt.xticks(rotation = 90)
plt.subplot(122)
plt.title('max of unitprice group by customerID')
plt.plot(max)
plt.xticks(rotation = 90)
plt.show()
```

```python
# 5. Find cumulative sales monthly and yearly and plot them
plt.title("sales monthly and yearly")
plt.ylabel("[CO2 Gas]")
plt.xlabel("Time (decade)")
plt.show()
```
