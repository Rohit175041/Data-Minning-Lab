```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# Load the dataset
df = pd.read_csv('studentData.csv')
df.head()
# df.describe()
```

```python
# 1. Find correlations between various features and grade_mean and see which features have a significant impact on it.

# Compute the mean of G1, G2, and G3
df['grade_mean'] = (df['G1'] + df['G2'] + df['G3']) / 3

# Compute the correlation matrix
corr_matrix = df.corr()

# Get the correlation of each feature with the grade_mean column
corr_with_grade_mean = corr_matrix['grade_mean'].sort_values(ascending=False)

# Print the correlations
print(corr_with_grade_mean)
```

```python
# 2. Find out got better grades (Boys or Girls). Similarly Find out who got better grades age wise.

df['mean'] = df[['G1', 'G2', 'G3']].mean(axis=1)

mean_grades_sex = df.groupby('sex')['mean'].mean()

bins = [0, 15, 17, 19, 21]
labels = ['<15', '15-17', '17-19', '19-21']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
mean_grades_age = df.groupby('age_group')['mean'].mean()


print("Mean Grades by Sex:")
print(mean_grades_sex)

print("\nMean Grades by Age Group:")
print(mean_grades_age)
```

```python
# 3. Count the number of distinct values in each column and drop column with very high distinct values.
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
    if df[col].nunique() > 60:
        df = df.drop(col,axis=1)
```

```python
# 4. Replace any missing value of each column by most frequent value having same Grade mean value.
# df.isnull().sum()
df.isnull().any()
# here is no any missing values
```

```python
#4 Lable encode all other columns.
le = preprocessing.LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
        
df.head(20)        
```
